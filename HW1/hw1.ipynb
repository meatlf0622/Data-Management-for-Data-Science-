```python
# FILL IN ALL THE FUNCTIONS IN THIS TEMPLATE
# MAKE SURE YOU TEST YOUR FUNCTIONS WITH MULTIPLE TEST CASES
# ASIDE FROM THE SAMPLE FILES PROVIDED TO YOU, TEST ON YOUR OWN FILES

# WHEN DONE, SUBMIT THIS FILE TO AUTOLAB

from collections import defaultdict
from collections import Counter

# YOU MAY NOT CODE ANY OTHER IMPORTS

# ------ TASK 1: READING DATA  --------

# 1.1
def read_ratings_data(f):
    # parameter f: movie ratings file name f (e.g. "movieRatingSample.txt")
    # return: dictionary that maps movie to ratings
    # WRITE YOUR CODE BELOW
    movieRatingDict = defaultdict(list)

    with open(f, 'r') as file:
        for line in file:
            movie, rating, userID = line.strip().split("|")
            movieRatingDict[movie].append(float(rating))

    return dict(movieRatingDict)

    pass
    

# 1.2
def read_movie_genre(f):
    # parameter f: movies genre file name f (e.g. "genreMovieSample.txt")
    # return: dictionary that maps movie to genre
    # WRITE YOUR CODE BELOW

    movieGenreDict = defaultdict(list)

    with open(f, 'r') as file: 
        for line in file:
            genre, movieID, nameYearMov = line.strip().split("|")
            movieGenreDict[nameYearMov] = genre
    return dict(movieGenreDict)

# ------ TASK 2: PROCESSING DATA --------

# 2.1
def create_genre_dict(d):
    # parameter d: dictionary that maps movie to genre
    # return: dictionary that maps genre to movies
    # WRITE YOUR CODE BELOW
    genreToMovDict = defaultdict(list)

    for movie,genre in d.items(): 
        genreToMovDict[genre].append(movie)
    return dict(genreToMovDict)
        
    
# 2.2
def calculate_average_rating(d):
    # parameter d: dictionary that maps movie to ratings
    # return: dictionary that maps movie to average rating
    # WRITE YOUR CODE BELOW
    avgRate = {}
    for movie, rating in d.items():
        avgRate[movie] = sum(rating) / len(rating)
    #maybe round to nearest 10th or 100th
    return dict(avgRate)
        
    
# ------ TASK 3: RECOMMENDATION --------

# 3.1
def get_popular_movies(d, n=10):
    # parameter d: dictionary that maps movie to average rating
    # parameter n: integer (for top n), default value 10
    # return: dictionary that maps movie to average rating, 
    #         in ranked order from highest to lowest average rating
    # WRITE YOUR CODE BELOW
    sortedMovies = sorted(d.items(), key = lambda x: (x[-1], x[0]))
    sortedMovies.reverse()
    #might have to change if the reverse is needed 
    return sortedMovies
    
# 3.2
def filter_movies(d, thres_rating=3):
    # parameter d: dictionary that maps movie to average rating
    # parameter thres_rating: threshold rating, default value 3
    # return: dictionary that maps movie to average rating
    # WRITE YOUR CODE BELOW
    filteredMovies = {}
    for movie, rating in d.items():
        if rating >= thres_rating: 
            filteredMovies[movie] = rating
    return filteredMovies
            
        
    pass
    
# 3.3
def get_popular_in_genre(genre, genre_to_movies, movie_to_average_rating, n=5):
    # parameter genre: genre name (e.g. "Comedy")
    # parameter genre_to_movies: dictionary that maps genre to movies
    # parameter movie_to_average_rating: dictionary  that maps movie to average rating
    # parameter n: integer (for top n), default value 5
    # return: dictionary that maps movie to average rating
    # WRITE YOUR CODE BELOW

    popInGenre = {}

    for movie in genre_to_movies.get(genre, []):
        if movie in movie_to_average_rating:
            popInGenre[movie] = movie_to_average_rating[movie]
    sortedMov = sorted(popInGenre.items(), key = lambda x: (-x[1],x[0]))
    
    return dict(sortedMov[:n])
    pass
    
# 3.4
def get_genre_rating(genre, genre_to_movies, movie_to_average_rating):
    # parameter genre: genre name (e.g. "Comedy")
    # parameter genre_to_movies: dictionary that maps genre to movies
    # parameter movie_to_average_rating: dictionary  that maps movie to average rating
    # return: average rating of movies in genre
    # WRITE YOUR CODE BELOW
    ratings = []
    for movie in genre_to_movies.get(genre, []):
        if movie in movie_to_average_rating:
            ratings.append(movie_to_average_rating[movie])
        return sum(ratings) / len(ratings) if ratings else 0

    pass
    
# 3.5
def genre_popularity(genre_to_movies, movie_to_average_rating, n=5):
    # parameter genre_to_movies: dictionary that maps genre to movies
    # parameter movie_to_average_rating: dictionary  that maps movie to average rating
    # parameter n: integer (for top n), default value 5
    # return: dictionary that maps genre to average rating
    # WRITE YOUR CODE BELOW
    genRate = {}
    for genre in genre_to_movies:
        genRate[genre] = get_genre_rating(genre, genre_to_movies, movie_to_average_rating)
    sorted_genres = sorted(genRate.items(), key=lambda x: (-x[1], x[0]))
    return dict(sorted_genres[:n])
    pass

# ------ TASK 4: USER FOCUSED  --------

# 4.1
def read_user_ratings(f):
    # parameter f: movie ratings file name (e.g. "movieRatingSample.txt")
    # return: dictionary that maps user to list of (movie,rating)
    # WRITE YOUR CODE BELOW
    userRating = defaultdict(list)
    with open(f, "r") as file:
        for line in file: 
            movie, rating, user = line.strip().split("|")
            userRating[user].append((movie, float(rating)))
    return dict(userRating)
    pass
    
# 4.2
def get_user_genre(user_id, user_to_movies, movie_to_genre):
    # parameter user_id: user id
    # parameter user_to_movies: dictionary that maps user to movies and ratings
    # parameter movie_to_genre: dictionary that maps movie to genre
    # return: top genre that user likes
    # WRITE YOUR CODE BELOW
    genreRate = defaultdict(list)
    for movie, rating in user_to_movies.get(user_id, []):
        genreRate[movie_to_genre.get(movie, "Unknown")].append(rating)
    avgGenRate = {g: sum(r) / len(r) for g, r in genreRate.items()}
    return max(avgGenRate , key = avgGenRate.get, default = None)
    pass
    
# 4.3    
def recommend_movies(user_id, user_to_movies, movie_to_genre, movie_to_average_rating):
    # parameter user_id: user id
    # parameter user_to_movies: dictionary that maps user to movies and ratings
    # parameter movie_to_genre: dictionary that maps movie to genre
    # parameter movie_to_average_rating: dictionary that maps movie to average rating
    # return: dictionary that maps movie to average rating
    # WRITE YOUR CODE BELOW
    if user_id not in user_to_movies:
        return {}
    
    user_movies = {m for m, _ in user_to_movies[user_id]}
    top_genre = get_user_genre(user_id, user_to_movies, movie_to_genre)
    if not top_genre:
        return {}
    
    recommendations = {m: movie_to_average_rating[m] for m in movie_to_genre if movie_to_genre[m] == top_genre and m not in user_movies and m in movie_to_average_rating}
    sorted_recommendations = sorted(recommendations.items(), key=lambda x: (-x[1], x[0]))
    
    return dict(sorted_recommendations[:3])

    pass

# -------- main function for your testing -----
def main():
    # write all your test code here
    # this function will be ignored by us when grading
    ratingFile = "rating.txt"
    genreFile = "genre.txt"

    #function: 1.1
    movieRate = read_ratings_data(ratingFile)
    print("Function 1.1")
    print(movieRate)
    
    #function: 1.2
    print("Function 1.2")
    genreMov = read_movie_genre(genreFile)
    print(genreMov)
    #function: 2.1 
    print("function 2.1")

    genDic = create_genre_dict(genreMov)
    print(genDic)
    #function: 2.2
    print("function 2.2")
    avgRatings = calculate_average_rating(movieRate)
    print(avgRatings)
    #function: 3.1
    print("fucntion 3.1")
    popMovies = get_popular_movies(avgRatings, 11)
    print(popMovies)
    #function: 3.2
    print("Function 3.2")
    filteredMovies = filter_movies(avgRatings, thres_rating = 3.5)
    print(filteredMovies)
    #function 3.3
    sortbyGenre = get_popular_in_genre("Comedy", genDic, avgRatings, n=3)
    print("function 3.3")
    print(sortbyGenre)
    #function 3.4 
    comRating = get_genre_rating("Comedy", genDic, avgRatings)
    print("Average Rating for Comedy Movies:")
    print(comRating)
    #function 3.5
    print("function 3.5)")
    genPop = genre_popularity(genDic, avgRatings, n=3)
    print("\nTop 3 Popular Genres:")
    print(genPop)
    #function 4.1 
    print("function 4.1")
    userRating = read_user_ratings(ratingFile)
    print(userRating)
    #function 4.2
    print("User 6, function 4.2")
    userGenre = get_user_genre("6", userRating, genreMov)
    print(userGenre)
    #function 4.3
    print("function 4.3")
    recommnedMov = recommend_movies("1", userRating, genreMov, avgRatings)
    print(recommnedMov)
    

# In other words, ALL code your write (including variable names) MUST be inside one of
# the above functions
    
# program will start at the following main() function call
# when you execute hw1.py
main()

    
```
