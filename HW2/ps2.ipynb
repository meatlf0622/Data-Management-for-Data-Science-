If not, increase the smaller number until the difference between the two numbers is equal to 5.
3. Return the small and large numbers in that order after performing the operation

---

```python
def foo(a, b):
    large, small = a,b 
    if a < b 
        large, small = b, a 
    dif = large - small
    if dif <= 5
        return small, large
    while large - small > 5:
        small += 1
    return small, large
```
